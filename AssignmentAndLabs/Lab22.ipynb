#### Exercise-4: 

You go to a market to pick out some gourds to decorate your house for Halloween. The market has a special October deal where you get 3 randomly chosen gourdss for $10. Each gourd can either be an orange pumpkin, a green pumpkin or a squash. The probabilities for picking each one of them are: P(orange pumpkin) = 0.6, P(green pumpkin) = 0.3, and P(squash) = 0.1.

1) What is the probability that the first gourd is an orange pumpkin, the second is a green pumpkin, and the third is a squash?

2) What is the probability that you get all 3 as orange pumpkins?

3) What is the probability that you get no squashes?

4) What is the probability that you get at least one orange pumpkin?

```python
#Solution:
# Given probabilities
p_orange = 0.6
p_green = 0.3
p_squash = 0.1

# Probability that the first gourd is an orange pumpkin, the second is a green pumpkin, and the third is a squash
p_1 = p_orange * p_green * p_squash

# Probability that you get all 3 as orange pumpkins
p_2 = p_orange ** 3

# Probability that you get no squashes
p_no_squash = 1 - p_squash
p_3 = p_no_squash ** 3

# Probability that you get at least one orange pumpkin
p_no_orange = 1 - p_orange
p_at_least_one_orange = 1 - (p_no_orange ** 3)

#results
print(f"Probability that the first gourd is an orange pumpkin, the second is a green pumpkin, and the third is a squash: {p_1:.4f}")
print(f"Probability that you get all 3 as orange pumpkins: {p_2:.4f}")
print(f"Probability that you get no squashes: {p_3:.4f}")
print(f"Probability that you get at least one orange pumpkin: {p_at_least_one_orange:.4f}")
```

#### Exercise-6: 

The outlet stream of a continuous chemical reactor is sampled every thirty minutes and titrated. Extensive records of normal operation show the concentration of component A obtained over 1000 batches is approximately normally distributed with mean 41.2 g/L and standard deviation 0.90 g/L.

1) Five determinations of concentration of component A are made. If the mean of these five concentrations is greater than or equal to 42.3 g/L, action is taken. What is the level of significance associated with this test?

2) State the null and the alternative hypothesis.

3) Give the statistical inference based on the conservative cut-off of 1%.

**Solution for 2): Defining null and alternative hypothesis:**

**Null hypothesis: Concentration of component A = 41.2 g/L**
    
**Alternative hypothesis: Concentration of component A != 41.2 g/L**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

conc = np.random.normal(41.2, 0.9, 1000)
plt.hist(conc, color = 'blue', ec = 'black')
plt.xlabel('Concentration of component A')
plt.ylabel('Count')
plt.title('Distribution of Concentration of component A')
```

```python
random_outcomes = np.random.choice(conc, 5)
sample_mean = np.mean(random_outcomes)
sample_mean
```

```python
repetitions = 10000

outcomes = []

for i in np.arange(10000):
    random_outcomes = np.random.choice(conc, 5)
    sample_mean = np.mean(random_outcomes)
    outcomes.append(sample_mean)
    
outcomes = np.array(outcomes)
```

```python
plt.hist(outcomes, color = 'red', ec = 'black')
plt.scatter(42.3, -60, color = 'blue')
plt.xlabel('Mean concentration of component A')
plt.ylabel('Sample count')
plt.title('Distribution of mean concentration of component A for 10,000 samples of size 5')
```

```python
#Solution for 1): Level of significance or p-value is:

p_value = np.count_nonzero(outcomes >= 42.3)/repetitions
p_value
```

**Solution for 3): Statistical inference:**

**The alternative hypothesis must be accepted where in the action must be taken because the p-value of approximately 0.23% is less than 1% which is highly statistically significant.**
