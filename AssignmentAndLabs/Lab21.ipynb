1. Read the `AirTraffic.csv` file as a dataframe and check its first few rows.<br>

3. Compute the arithmetic and harmonic mean of `Distance`.<br>

5. Find the range of `Distance`.<br>

7. Use descriptive functions of the Pandas library to get a 5-number summary of 'Distance'. Plot a box plot without outliers.<br>

```python
AT = pd.read_csv("AirTraffic.csv")         #Read the .csv file a data frame
AT.head()
```

```python
AT.info()
```

```python
Distance = AT['DISTANCE']
# Use the mean function from the statistics library
mean = statistics.mean(Distance)
print("The arithmetic mean distance of the 2020 flights is ",round(mean,2),"miles")
hmean = statistics.harmonic_mean(Distance)
print("The harmonic mean distance of the 2020 flights is ",round(hmean,2),"miles")
```

```python
Distance = AT['DISTANCE']
# Use the mean function from the statistics library
median = statistics.median(Distance)
print("The median of distance of the 2020  flights is ",median,"miles")
```

```python
Distance = AT['DISTANCE']
Range = np.ptp(Distance) #ptp stands for Peak To Peak
print("The range of distance of the 2020  flights is ",Range,"miles")
```

```python
Distance = AT['DISTANCE']

IQR = scipy.stats.iqr(Distance)
print("The IQR of distance of the 2020  flights is ",IQR,"miles")
```

```python
Distance = AT['DISTANCE']
Distance.describe()
```

```python
fig = plt.figure(figsize =(7, 5))
plt.boxplot(Distance,medianprops={'linewidth': 1, 'color': 'purple'},showfliers=False)

plt.show()
```

```python
Distance = AT['DISTANCE']
var = statistics.variance(Distance)
sd = statistics.stdev(Distance)
print("The variance and standard deviation of distance of the 2020 flights is ",round(var,2)," and ",round(sd,2)," respectively")
```

```python
Distance = AT['DISTANCE']
skew = scipy.stats.skew(Distance)
kurtosis = scipy.stats.kurtosis(Distance)
print("The skewness and kurtosis of distance of the 2020 flights is ",round(skew,2)," and ",round(kurtosis,2)," respectively")
```

1. Read the "Lubbock_Oct_T&P.csv" file as a dataframe and check its first few rows.<br>

6. Find the range and IQR of 'precipitation'.<br>

```python
#code here
# read the data into a dataframe
# info about the dataframe
# descriptive statistics for temperature
# median of precipitation and temperature
# how many modes temperature
# IQR precipitation
# Selected quantiles for temperature
# 5-number summary precipitation
# Boxplot
# Variance and standard deviation of 'precipitation'
# Skewness and kurtosis 'precipitation'

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("Lubbock_Oct_T&P.csv")
print("DataFrame:")
print(df.head())

#  format of the dataframe
print("DataFrame Information:")
print(df.info())
print("Summary Statistics:")
print(df.describe())


# arithmetic, harmonic, and geometric mean
arithmetic_mean_temp = df['temperature'].mean()
harmonic_mean_temp = len(df['temperature']) / (1 / df['temperature']).sum()
geometric_mean_temp = np.exp(np.mean(np.log(df['temperature'])))
print(f"Arithmetic Mean of Temperature: {arithmetic_mean_temp} °C")
print(f"Harmonic Mean of Temperature: {harmonic_mean_temp} °C")
print(f"Geometric Mean of Temperature: {geometric_mean_temp} °C")

# median 
median_precipitation = df['precipitation'].median()
median_temperature = df['temperature'].median()
print(f"Median Precipitation: {median_precipitation} mm")
print(f"Median Temperature: {median_temperature} °C")


# one mode, two modes, or multiple modes.
temperature_modes = df['temperature'].mode()
num_modes = len(temperature_modes)
if num_modes == 1:
    modes_description = "The set of 'temperature' has one mode."
elif num_modes == 2:
    modes_description = "The set of 'temperature' has two modes."
else:
    modes_description = "The set of 'temperature' has multiple modes."
print(modes_description)


# range and IQR
precipitation_range = df['precipitation'].max() - df['precipitation'].min()
Q1 = df['precipitation'].quantile(0.25)
Q3 = df['precipitation'].quantile(0.75)
precipitation_IQR = Q3 - Q1
print(f"Range of Precipitation: {precipitation_range} mm")
print(f"IQR of Precipitation: {precipitation_IQR} mm")
```
