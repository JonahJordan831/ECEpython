### Step6: Get the sodium column.

```python
sodium = nutrition_df[" Sodium (mg)"]
```

### Step7: Plot a histogram of sodium content.

```python
fig = plt.figure(figsize = (15, 5)) 
plt.hist(sodium, color ='darkgreen')
plt.title("Sodium in Starbucks Menu Items")
```

### Step8: What are the unique categories of beverages?

```python
nutrition_df['Beverage_category'].unique()
```

### Step9: How many calories does each category of beverage have?

```python
#Pull two columns "Beverage_category" and "Calories" from df
#Rename these two columns to "DrinkMenu" and "Calories"

calories_df = pd.DataFrame()

calories_df = nutrition_df[['Beverage_category','Calories']]
calories_df.columns = ['Category', 'Calories']
calories_df
```

```python
beverage = list(calories_df['Category'])
calories = list(calories_df['Calories'])
   
fig = plt.figure(figsize = (15, 10)) 

plt.barh(beverage, calories, color ='darkred',  height = 0.625) 
  
plt.xlabel("Calories") 
plt.ylabel("Beverage Category") 
plt.title("Calories by Category of Beverage") 
plt.show() 
```

```python
calories_df.groupby('Category').mean()
```

### Step9 (***corrected***): How many calories does each category of beverage have?

```python
# numberOfCategories = calories_df['Category'].nunique()
# numberOfCategories

mean_cal_df = pd.DataFrame()
```

```python
mean_cal_df = calories_df.groupby('Category').mean()
mean_cal_df.reset_index(inplace=True)
```

```python
# beverage = list(calories_df['DrinkMenu'].unique())
beverage = list(mean_cal_df['Category'])
calories = list(mean_cal_df['Calories'])

fig = plt.figure(figsize = (15, 10)) 

plt.barh(beverage, calories, color ='darkgreen',  height = 0.625) 
  
plt.xlabel("Calories") 
plt.ylabel("Beverage Category") 
plt.title("Calories by Category of Beverage") 
plt.show() 
```

### Step10: Which Starbucks drink category has the highess calories?

```python
mean_cal_df.sort_values("Calories", ascending=True, inplace=True)
```

```python
beverage = list(mean_cal_df['Category'])
calories = list(mean_cal_df['Calories'])

fig = plt.figure(figsize = (15, 10)) 

plt.barh(beverage, calories, color ='saddlebrown',  height = 0.625) 
  
plt.xlabel("Calories") 
plt.ylabel("Beverage Category") 
plt.title("Calories by Category of Beverage") 
plt.show() 
```

### Step11: Plot a historgram of calories data.

```python
fig = plt.figure(figsize = (10, 5)) 

calories = nutrition_df["Calories"]
plt.hist(calories, bins=10, edgecolor = "black", color = "darkgreen")
plt.title("Calories in Starbucks Menu Items")
plt.xlabel("Calories in kcal")
plt.ylabel("Count")
```

### Step12: Observe the relation between Calories and Beverage Preperation method using a scatter plot.

```python
fig = plt.figure(figsize = (10, 5)) 

plt.scatter(nutrition_df["Calories"],nutrition_df['Beverage_prep'], color ='darkgreen')

plt.title('Scatter Plot of Calories Relation with Beverage Preperation') # Set the plot title
plt.xlabel('Calories')                           # Set the label for x-axis
plt.ylabel('Preperation')                     # Set the label for y-axis
```
