```python
# Read the .csv file a data frame
df = pd.read_csv("AirTraffic.csv")

# Display the first few rows of the DataFrame
df.head()
```

```python
# info
```

```python
# Use the mean function from the statistics library
# Use the harmonic mean function from the statistics library
df.info()


df.describe()

harmonic_mean = len(df['DISTANCE']) / (1 / df['DISTANCE']).sum()

harmonic_mean
```

```python
# Use the median function from the statistics library
median_distance = df['DISTANCE'].median()
print("the median distance is", median_distance)
```

```python
# Peak To Peak
distance_range = df['DISTANCE'].max() - df['DISTANCE'].min()
distance_range
print("the range is", distance_range)
```

```python
# IQR
Q1 = df['DISTANCE'].quantile(0.25)
Q3 = df['DISTANCE'].quantile(0.75)
IQR = Q3 - Q1
print("theIQR is", IQR)
```

```python
# describe
summary = df['DISTANCE'].describe()




summary
```

```python
# box plot
df.boxplot(column='DISTANCE', showfliers=False)  # Set showfliers=False to exclude outliers
plt.title("Box Plot of Distance (Without Outliers)")
plt.ylabel("DISTANCE")
plt.show()
```

```python
# variance
# stdev
variance_distance = df['DISTANCE'].var()

# Find the standard deviation of 'Distance'
std_dev_distance = df['DISTANCE'].std()

print("the variencce and stdev is", variance_distance, "and", std_dev_distance)
```

```python
# skew
# kurtosis
# Find the skewness of 'Distance'
skewness_distance = df['DISTANCE'].skew()

# Find the kurtosis of 'Distance'
kurtosis_distance = df['DISTANCE'].kurtosis()
 
print("the variencce and stdev is", skewness_distance, "and", kurtosis_distance)
```
