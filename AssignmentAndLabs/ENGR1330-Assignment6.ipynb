**Task-3:** Our focus is now on the compressive strength that is located in the **`csMPa`** column. Compute the mean and the standard deviation of the column **`csMPa`** using the appropriate functions available in the Pandas library.

**Task-4:** Plot an histogram of the distribution of the compressive strength values in MPa that are located in the **`csMPa`** column. For the bins, you must choose the reasonable start point and the end point by looking at the minimum and the maximum value of the **`csMPa`** column. Further, the bin-width must be fixed depending on the range of the **`csMPa`** column and the number of bins. Also, the histogram plot must include appropriate xlabel, ylabel, and title.

**Note:** You have to refer to the matplotlib lesson slides to determine the bin-width based on the values of the range and the number of bins.

```python
#GIVE YOUR ANSWER FOR TASK-4 IN THIS CELL
min_csMPa = concrete['csMPa'].min()
max_csMPa = concrete['csMPa'].max()

range_csMPa = max_csMPa - min_csMPa
num_bins = int(range_csMPa / 5) 

#histogram
plt.hist(concrete['csMPa'], bins=num_bins, range=(min_csMPa, max_csMPa), edgecolor='black')
plt.xlabel('Compressive Strength (csMPa)')
plt.ylabel('Frequency')
plt.title('Distribution of Compressive Strength')
plt.grid(axis='y', alpha=0.75)

plt.show()
```

**Task-5:** Pick a random sample of size 100 (without replacement) from the dataframe **`concrete`** and store the sampled rows in a new variable name **`concrete_sample_100`**.  Plot an histogram of the distribution of the compressive strength values that are located in the **`csMPa`** column of the **`concrete_sample_100`** dataframe. Use the same start point, end point, and bin size that you used for the previous task. Also, the histogram plot must include appropriate xlabel, ylabel, and title.

**Note:** Use the **`sample()`** function in the Pandas library.

```python
#GIVE YOUR ANSWER FOR TASK-5 IN THIS CELL
concrete_sample_100 = concrete.sample(n=100, replace= False)  

# min and max
min_csMPa = concrete['csMPa'].min()
max_csMPa = concrete['csMPa'].max()

#  range and number of bins
range_csMPa = max_csMPa - min_csMPa
num_bins = int(range_csMPa / 5)

# histogram 
plt.hist(concrete_sample_100['csMPa'], bins=num_bins, range=(min_csMPa, max_csMPa), edgecolor='black')
plt.xlabel('Compressive Strength (csMPa)')
plt.ylabel('Frequency')
plt.title('Distribution of Compressive Strength in Sample of 100')
plt.grid(axis='y', alpha=0.75)

plt.show()
```

**Task-6:** Pick a random sample of size 400 (without replacement) from the dataframe **`concrete`** and store the sampled rows in a new variable name **`concrete_sample_400`**.
```python
#GIVE YOUR ANSWER FOR TASK-6 IN THIS CELL
concrete_sample_400 = concrete.sample(n=400, replace= False)  

# min and max
min_csMPa = concrete['csMPa'].min()
max_csMPa = concrete['csMPa'].max()

#  range and number of bins
range_csMPa = max_csMPa - min_csMPa
num_bins = int(range_csMPa / 5)

# histogram 
plt.hist(concrete_sample_400['csMPa'], bins=num_bins, range=(min_csMPa, max_csMPa), edgecolor='black')
plt.xlabel('Compressive Strength (csMPa)')
plt.ylabel('Frequency')
plt.title('Distribution of Compressive Strength in Sample of 400')
plt.grid(axis='y', alpha=0.75)

plt.show()
```

**Task-7:** In order to convince ourselves that the sample size of 400 is better compared to 100, we have to perform some simulation to compute sample statistics. In this task, pick 10,000 random samples of size 100 (without replacement) from the dataframe **`concrete`** and store it in a variable name **`concrete_sample_100`** inside the for loop. Then, find the mean of **`csMPa`** column located in the dataframe **`concrete_sample_100`** and store it in a variable name **`mean`** inside the for loop. Append the mean that is calculated for each of the 10,000 random samples to the list named **`sample100_cs_mean`**. Plot an histogram to visualize the sampling distribution of **`sample100_cs_mean`** and also, plot the **`pop_cs_mean`** (calculated in task-3) to see where it lies in the sampling distribution of **`sample100_cs_mean`** using the **`scatter()`** function.   

```python
#COMPLETE YOUR ANSWER FOR TASK-7 IN THIS CELL

repetitions = 10000


sample100_cs_mean = []

for i in range(repetitions):
   
    concrete_sample_100 = concrete.sample(n=100, replace=False)
    
  
    mean = concrete_sample_100['csMPa'].mean()
    
    # Append
    sample100_cs_mean.append(mean)

sample100_cs_mean = np.array(sample100_cs_mean)

# Plot a histogram to visualize the sampling distribution of sample means
plt.hist(sample100_cs_mean, bins=30, edgecolor='black', alpha=0.7, label='Sample Means')
plt.axvline(pop_cs_mean, color='red', linestyle='dashed', linewidth=2, label='Population Mean')
plt.xlabel('Sample means for sample size 100')
plt.ylabel('Count of sample means')
plt.title('Sampling distribution of mean compressive strength for 10,000 samples of size 100')
plt.legend()
plt.show()




```
